```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Billionaire.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
data.isnull().sum()
```

```python
data["NetWorth"] = data["NetWorth"].str.strip("B")
data["NetWorth"] = data["NetWorth"].str.strip("$")
data["NetWorth"] = data["NetWorth"].astype("float")
```

```python
data.head()
```

```python
df = data.sort_values(by = ["NetWorth"], ascending = False)
df
```

```python
df = data.sort_values(by = ["NetWorth"], ascending = False).head(10).reset_index()
df
```

```python
plt.figure(figsize=(18,6))
plt.bar(df["Name"], df["NetWorth"])
plt.title("Top 10 the most NetWorth people in the world ")
plt.xlabel("People")
plt.ylabel("NetWorth($B)")
plt.grid()
plt.show()
```

```python
df = data["Source"].value_counts().head(10).reset_index()
df
```

```python
myexplode = [0.04, 0.04, 0.04, 0.04, 0.04, 0.04, 0.04, 0.04, 0.04, 0.04]
plt.pie(df["count"], labels = df["Source"], explode = myexplode, shadow = True)
plt.title("Top 10 Sources of Making Money")
plt.show()
```

```python
df = data["Industry"].value_counts().reset_index()
df
```

```python
plt.figure(figsize=(24,12))
plt.pie(df["count"], labels = df["Industry"])
plt.title("Top Industry of Making Money")
plt.show()
```

```python
df = data["Country"].value_counts().head(5).reset_index()
df
```

```python
myexplode = [0.04, 0.04, 0.04, 0.04, 0.04]
plt.pie(df["count"], labels = df["Country"], explode = myexplode, shadow = True)
plt.title("Top 5 Countries of Making Money")
plt.show()
```

```python
df = data["Age"].value_counts().reset_index().sort_values(by="Age", ascending=False)
df
```

Average age in Billionaire people

```python
average_age = df["Age"]
average_age = np.mean(average_age)
average_age
```

Variance age in Billionaire people

```python
var_age = df["Age"]
var_age = np.var(var_age)
var_age
```

```python
country_networth = data.groupby('Country')['NetWorth'].sum().reset_index().sort_values(by="NetWorth", ascending=False).head(10)
country_networth
```

```python
myexplode = [0.04, 0.04, 0.04, 0.04, 0.04, 0.04, 0.04, 0.04, 0.04, 0.04]
plt.figure(figsize=(24,8))
plt.pie(country_networth["NetWorth"], labels= country_networth["Country"], autopct='%1.1f%%', explode = myexplode, shadow = True)
plt.title("Percent of the most NetWorth in top 10 countries")
plt.show()
```

```python
plt.figure(figsize=(24,10))
plt.bar(country_networth["Country"], country_networth["NetWorth"])
plt.title("NetWorth of top 10 countries")
plt.xlabel("countries")
plt.ylabel("NetWorth($B)")
plt.grid()
plt.show()
```

```python
people_networth = data.groupby('Name')['NetWorth'].sum().reset_index().sort_values(by="NetWorth", ascending=False).head(10)
people_networth
```

```python
plt.figure(figsize=(24,10))
plt.bar(people_networth["Name"], people_networth["NetWorth"])
plt.title("Top 10 the most NetWorth people in the world ")
plt.xlabel("People")
plt.ylabel("NetWorth($B)")
plt.grid()
plt.show()
```
