```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from perceptron import Perceptron
```

```python
X, Y = make_regression(n_samples=1000, n_features=1, noise=10, random_state=42)
scaler_X = MinMaxScaler(feature_range=(0, 14))
X = scaler_X.fit_transform(X)
scaler_Y = MinMaxScaler(feature_range=(0, 100))
Y = scaler_Y.fit_transform(Y.reshape(-1, 1)).flatten()
```

```python
plt.scatter(X, Y, color = "blue", marker = ".", label="Employee")
plt.title("Employee's salary")
plt.xlabel("Time (Hours)")
plt.ylabel("Salary (Million Toman)")
plt.legend()
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
perceptron = Perceptron(input_size=X_train.shape[1], lr=0.001, epochs=100)
perceptron.fit(X_train, Y_train)
```

```python
perceptron = Perceptron(input_size=X_train.shape[1], lr=0.001, epochs=100)
perceptron.fit(X_train, Y_train)
```

```python
mse = perceptron.evaluate(X_test, Y_test)
print("Mean Squared Error on Test Data (MSE):", mse)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].scatter(X, Y, color="blue", label="Employee")
axs[0].plot(X, perceptron.predict(X), color="red", label="Model Prediction")
axs[0].set_title("Employee's Salary Prediction")
axs[0].set_xlabel("Time (Hours)")
axs[0].set_ylabel("Million Toman")
axs[0].legend()

axs[1].plot(range(perceptron.epochs), perceptron.loss_history, color="green")
axs[1].set_title("Loss over Epochs")
axs[1].set_xlabel("Epoch")
axs[1].set_ylabel("Mean Squared Error (Loss)")
plt.tight_layout()
plt.show()
```
