```python
import numpy as np
from knn import KNN
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
nemo = cv2.imread("nemo0.jpg")
nemo = cv2.resize(nemo, (0,0), fx=0.5, fy=0.5)
nemo = cv2.cvtColor(nemo, cv2.COLOR_BGR2RGB)
plt.imshow(nemo)
plt.show()
```

```python
pixels_list_rgb = nemo.reshape(-1, 3) # we dont know how many row = -1
pixels_list_rgb.shape
fig = plt.figure(figsize=(14,16))
ax = fig.add_subplot(projection = "3d")
ax.scatter(pixels_list_rgb[:, 0], pixels_list_rgb[:, 1], pixels_list_rgb[:, 2], c=pixels_list_rgb / 255)
ax.set_xlabel("Red")
ax.set_ylabel("Blue")
ax.set_zlabel("Green")
plt.show()
```

```python
nemo_hsv = cv2.cvtColor(nemo, cv2.COLOR_RGB2HSV)
```

```python
pixels_list_hsv = nemo_hsv.reshape(-1, 3)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection = "3d")
ax.scatter(pixels_list_hsv[:, 0], pixels_list_hsv[:, 1], pixels_list_hsv[:, 2], c=pixels_list_rgb / 255)
ax.set_xlabel("Hue")
ax.set_ylabel("Saturation")
ax.set_zlabel("Value")
plt.show()
```

```python
light_orange = (1, 190, 200)
dark_orange = (18, 255, 255)

mask = cv2.inRange(nemo_hsv, light_orange, dark_orange)
plt.imshow(mask, cmap="gray")
plt.show()
```

```python
light_white = (0, 0, 200)
dark_white = (145, 60, 255)

mask_white = cv2.inRange(nemo_hsv, light_white, dark_white)
plt.imshow(mask_white, cmap="gray")
plt.show()
```

```python
final_mask = mask + mask_white
plt.imshow(final_mask, cmap="gray")
plt.show()
```

```python
final_result = cv2.bitwise_and(nemo, nemo, mask=final_mask)
plt.imshow(final_result)
plt.show()
```

```python
X_train = pixels_list_hsv / 255
X_train.shape
```

```python
Y_train = final_mask.reshape(-1,) //255
Y_train.shape
```

```python
Y_train
```

```python
knn = KNN(k=3)
knn.fit(X_train, Y_train)
```

```python
new_pixel = np.array([10, 200, 255]) / 255
knn.predict([new_pixel])
```
