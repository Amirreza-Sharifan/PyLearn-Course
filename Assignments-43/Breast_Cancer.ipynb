```python
import numpy as np
from knn import KNN
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
cancer = load_breast_cancer()
X = cancer.data
Y = cancer.target

X_train , X_test, Y_tain, Y_test = train_test_split(X, Y, test_size=0.2)
```

KNN-3

```python
knn_Amirreza_3 = KNN(k=3)
knn_Amirreza_3.fit(X_train, Y_tain)
accuracy_a_3 = knn_Amirreza_3.evaluate(X_test, Y_test)
print("Accuracy of KNN(3)_Amirreza", accuracy_a_3)
    
knn_sklearn_3 = KNeighborsClassifier(n_neighbors=3)
knn_sklearn_3.fit(X_train, Y_tain)
accuracy_s_3 = knn_sklearn_3.score(X_test, Y_test)
print("Accuracy of KNN(3)_sklearn:", accuracy_s_3)
```

```python
Y_predict = knn_sklearn_3.predict(X_test)
conf_matrix = confusion_matrix(Y_test, Y_predict)
conf_matrix
```

```python
plt.imshow(conf_matrix, cmap='Blues')
plt.title('Confusion Matrix (K=3)')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.colorbar()

classes = cancer.target_names
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)

for i in range(len(classes)):
    for j in range(len(classes)):
        plt.text(j, i, conf_matrix[i, j], ha='center', va='center', color='red')

plt.show()
```
