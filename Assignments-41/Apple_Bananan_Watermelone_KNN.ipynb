```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 200
apple_width = np.random.normal(6, 1, n)
apple_height = np.random.normal(6, 1, n)
banana_width = np.random.normal(4, 1, n)
banana_height = np.random.normal(8, 1, n)
watermelon_width = np.random.normal(6, 1, n)
watermelon_height = np.random.normal(10, 1, n)
```

```python
plt.scatter(apple_width, apple_height, label = "Apple", color = "red", marker = ".")
plt.scatter(banana_width, banana_height, label = "Banana", color = "orange", marker = ".")
plt.scatter(watermelon_width, watermelon_height, label = "Watermelon", color = "green", marker = ".")
plt.title("Apple & Banana normal distribution")
plt.xlabel("width")
plt.ylabel("height")
plt.legend()
plt.show()
```

KNN = K Nearest Neighbors

```python
width = np.concatenate((apple_width, banana_width, watermelon_width))
height = np.concatenate((apple_height, banana_height, watermelon_height))
X_train = np.array([width ,height]).T
```

```python
Y_apple = np.zeros(n, dtype=int)
Y_banana = np.ones(n, dtype=int)
Y_watermelon = np.ones(n, dtype=int) * 2
Y_trian = np.concatenate([Y_apple, Y_banana, Y_watermelon])
```

```python
def oq_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))
```

```python
new_fruit = np.array([6,9])

distance = []
for x in X_train:
    d = oq_distance(new_fruit, x)
    distance.append(d)
```

```python
k=5
```

```python
nearest_neighbor = np.argsort(distance)[0:k]
```

```python
Y_trian[nearest_neighbor]
```

```python
result = np.bincount(Y_trian[nearest_neighbor])
result
```

```python
final = np.argmax(result)
```

```python
if final == 0:
    print("🍎")
elif final == 1:
    print("🍌")
else:
    print("🍉")
```
