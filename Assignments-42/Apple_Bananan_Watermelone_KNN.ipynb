```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
def generait_data(n):
    apple_width = np.random.normal(6, 1, n)
    apple_height = np.random.normal(6, 1, n)
    banana_width = np.random.normal(4, 1, n)
    banana_height = np.random.normal(8, 1, n)
    watermelon_width = np.random.normal(6, 1, n)
    watermelon_height = np.random.normal(10, 1, n)

    width = np.concatenate((apple_width, banana_width, watermelon_width))
    height = np.concatenate((apple_height, banana_height, watermelon_height))
    X = np.array([width ,height]).T
    
    Y_apple = np.zeros(n, dtype=int)
    Y_banana = np.ones(n, dtype=int)
    Y_watermelon = np.ones(n, dtype=int) * 2
    Y = np.concatenate([Y_apple, Y_banana, Y_watermelon])

    return X,Y
    
```

```python
X_train, Y_trian = generait_data(100)
```

```python
plt.scatter(X_train[0:100, 0], X_train[0:100, 1],label = "Apple", color = "red", marker = ".")
plt.scatter(X_train[101:200, 0], X_train[101:200, 1], label = "Banana", color = "orange", marker = ".")
plt.scatter(X_train[201:300, 0], X_train[201:300, 1], label = "Watermelon", color = "green", marker = ".")
plt.title("Apple, Banana & Watermelon normal distribution")
plt.xlabel("width")
plt.ylabel("height")
plt.legend()
plt.grid()
plt.show()
```

KNN = K Nearest Neighbors

```python
from knn import KNN
```

```python
knn = KNN(5)
knn.fit(X_train, Y_trian)
```

```python
new_fruit = np.array([3,5])
new_fruit_2 = np.array([4,7])
new_fruit_3 = np.array([10,8])

new_fruits = [new_fruit, new_fruit_2, new_fruit_3]
```

```python
final = knn.predict(new_fruits)
```

```python
for x in final:
    if x == 0:
        print("üçé")
    elif x == 1:
        print("üçå")
    else:
        print("üçâ")
```

```python
n = 90
X_test , Y_test = generait_data(n)
```

```python
knn.evaluate(X_test, Y_test)
```

```python
Y_predict = knn.predict(X_test)
```

```python
confusion_matrix = np.zeros((3,3))

for i in range (3 * n):
    if Y_test[i] == 0 and Y_predict[i] == 0:
        confusion_matrix[0][0] += 1
    elif Y_test[i] == 0 and Y_predict[i] == 1:
        confusion_matrix[0][1] += 1
    elif Y_test[i] == 0 and Y_predict[i] == 2:
        confusion_matrix[0][2] += 1
    elif Y_test[i] == 1 and Y_predict[i] == 0:
        confusion_matrix[1][0] += 1
    elif Y_test[i] == 1 and Y_predict[i] == 1:
        confusion_matrix[1][1] += 1
    elif Y_test[i] == 1 and Y_predict[i] == 2:
        confusion_matrix[1][2] += 1
    elif Y_test[i] == 2 and Y_predict[i] == 0:
        confusion_matrix[2][0] += 1
    elif Y_test[i] == 2 and Y_predict[i] == 1:
        confusion_matrix[2][1] += 1
    elif Y_test[i] == 2 and Y_predict[i] == 2:
        confusion_matrix[2][2] += 1

confusion_matrix
```

```python
plt.imshow(confusion_matrix)
plt.colorbar()
plt.show()
```
